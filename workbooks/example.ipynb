## Visualizing Data

### Line Plot

We can visualize data using line plots. For example, let's plot a simple linear function:

```python
x = np.linspace(0, 10, 100)
y = 2 * x + 3

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Linear Function: y = 2x + 3')
plt.grid(True)
plt.show()
```

This code generates a line plot of the linear function `y = 2x + 3`.
